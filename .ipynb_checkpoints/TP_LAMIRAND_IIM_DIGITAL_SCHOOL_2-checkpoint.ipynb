## Importing Librairies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate, cross_val_score
```

## Preprocessing

- Check if missing values

- Check if duplicates ( if yes delete duplicated rows)

```python
# Load the CSV file
data = pd.read_csv('./data/train.csv')

# Check for missing values
missing_values = data.isnull().sum().sum()

if missing_values > 0:
    print("The CSV file contains missing values.")
else:
    print("The CSV file does not contain any missing values.")
```

```python
# Check for duplicates
duplicates = data.duplicated()
if duplicates.any():
    print("The CSV file contains duplicates. Removing them...")
    data = data.drop_duplicates()
    print("Duplicates removed. Updated DataFrame:\n", data)
else:
    print("The CSV file does not contain any duplicates.")
```
